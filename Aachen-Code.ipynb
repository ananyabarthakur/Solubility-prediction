```python
def Lasso():
    #USING LASSO
    from sklearn.linear_model import Lasso
    from sklearn.feature_selection import SelectFromModel
   
    lasso = Lasso(alpha=0.01,max_iter=10000, random_state=1).fit(X_train, y_train)

    model = SelectFromModel(lasso, prefit=True)
    X_new_lasso = model.transform(X_train)

    # Get back the kept features as a DataFrame with dropped columns as all 0s
    selected_features = pd.DataFrame(model.inverse_transform(X_new_lasso),
                                     index= X_train.index, 
                                     columns= X_train.columns)
    # Dropped columns have values of all 0s, keep other columns 
    selected_columns_lasso = selected_features.columns[selected_features.var() != 0]

    selected_data_train = X_train[selected_columns_lasso]             
    selected_data_test = X_test[selected_columns_lasso]
    selected_data_test = selected_data_test.apply(pd.to_numeric)

    #print(selected_data_train.columns)
    #print("Total selected descriptors by LASSO:",len(selected_data_train.columns))
    return selected_data_train, selected_data_test
```

```python
def prediction(X_train, X_test, y_train, y_test, model):
    if (model == 'RF'): 
        model = RandomForestRegressor(random_state = 0, n_estimators=100)
    elif model == 'Ridge':
        model = Ridge(alpha=1.0)
    elif model == 'ANN':
        model = MLPRegressor(activation='tanh', hidden_layer_sizes=(500), max_iter=500, random_state=0, solver='adam')
    else:
        print ('invalid')
    print ('starting...')
    model.fit(X_train, y_train) 
    pred = model.predict(X_test)
    
    print ("--model performance--")
    rmse = (np.sqrt(mean_squared_error(y_test, pred)))
    r2 = r2_score(y_test, pred)
    mae = MAE (y_test, pred)
    print('RMSE is {}'.format(rmse))
    print('R2 score is {}'.format(r2))
    print('MAE is {}'.format(mae))
    print('\n')
```

```python
possible_opts = ['none', 'ff', 'pm7', 'bad-data']
```

```python
def gen_all_predictions ():
    print ('Max possible datasets: ')
    for opt in possible_opts:
        if opt == 'bad-data':
            continue
        print ('---- '+ opt +' ----')
        X_train, X_test, y_train, y_test = load_data(opt, True, False)
        print ('X_train shape:', X_train.shape)
        prediction (X_train, X_test, y_train, y_test, 'RF' )
    
    
        
    print ('Same size datasets, 2D and 3D:  ')
    for opt in possible_opts:
        print ('---- '+ opt +' ----')
        X_train, X_test, y_train, y_test = load_data(opt, True, True)
        print ('X_train shape:', X_train.shape)
        prediction (X_train, X_test, y_train, y_test, 'RF' )

    
  
```
